4. **Обработка выбросов**: [+1 балл]
   - Определить пороги для выбросов на основе статистических мер (например, IQR)
   - Решить, удалять ли выбросы или корректировать их значения.

````python
df.head()
````

Считаем межквартильные размахи для численных типов данных:

````python
index_iqr = df.columns
index_iqr = index_iqr.drop(['CHAS', 'CHAS_bool', 'RAD'])
q1 = df[index_iqr].quantile(0.25)
q3 = df[index_iqr].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
df_iqr = pd.DataFrame({'q1':q1, 'q3':q3, 'iqr':iqr, 'lower':lower_bound, 'upper':upper_bound})
df_iqr.head()
````

Фильтруем исходные данные:

````python
result = list()
for item in index_iqr:
    tmp = df[item]
    tmp = tmp[(tmp >= df_iqr['lower'][item]) & (tmp <= df_iqr['upper'][item])]
    result.append(tmp)
filtered_df = pd.DataFrame()
for index, item in enumerate(result):
    filtered_df.insert(index, item.name, item)
filtered_df.head()
````

### Что такое IQR и как его использовать для обработки выбросов:

**IQR (Interquartile Range)**, или межквартильный размах, это мера статистического разброса и используется для определения выбросов. IQR рассчитывается как разность между 75-м и 25-м процентилями данных (Q3 и Q1 соответственно).

Выбросы часто определяются как значения, которые находятся за пределами следующих границ:
- Нижняя граница: Q1 - 1.5 * IQR
- Верхняя граница: Q3 + 1.5 * IQR

#### Примерный код на Python с использованием pandas:

```python
import pandas as pd

# Загрузка данных
df = pd.read_csv('path_to_file.csv')

# Рассчитываем IQR для колонки 'example_column'
Q1 = df['example_column'].quantile(0.25)
Q3 = df['example_column'].quantile(0.75)
IQR = Q3 - Q1

# Определяем границы для выбросов
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Фильтрация данных для удаления или корректировки выбросов
filtered_df = df[(df['example_column'] >= lower_bound) & (df['example_column'] <= upper_bound)]
```
